```python
t = pandas.read_csv('quiz3.txt', delimiter='\n', names=['input'])
```

```python
id1 = '1 @ 1,3: 4x4'
id2 = '2 @ 3,1: 4x4'
id3 = '3 @ 5,5: 2x2'
```

```python
def get_corners(string):
    string = string.split('@')[1]
    
    t = dict()
    
    t['left_edge'] = int(string.split(':')[0].split(',')[0])
    t['top_edge'] = int(string.split(':')[0].split(',')[1])
    
    width = int(string.split(':')[1].split('x')[0])
    height = int(string.split(':')[1].split('x')[1])
    
    t['right_edge'] = t['left_edge'] + width
    t['bottom_edge'] = t['top_edge'] + height
    
    return t

get_corners(id1)
get_corners(t['input'][0])
```

```python
drawing = [['.']*2000]*2000
```

```python
def get_drawing(lines):
    drawing = [['.']*2000]*2000
    
    id = 1
    for line in lines:
        corners = get_corners(line)
        print(corners)
        for l in range(corners['top_edge'], corners['bottom_edge']+1):
            for k in range(corners['left_edge'], corners['right_edge']+1):
                
                if drawing[l][k] == '.':
                    print('hi')
                    drawing[l][k] = str(id)
                else:
                    drawing[l][k] = 'X'
        id+=1
    return drawing
```

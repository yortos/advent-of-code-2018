```python
def power_level(x,y, serial_number= 4151):
    rack_id = x+10
    power_level = rack_id*y
    power_level+=serial_number
    power_level = power_level*rack_id
    
    hundred_digit = str(power_level)
    if len(hundred_digit)>=3:
        hundred_digit = int(hundred_digit[-3])
    else:
        hundred_digit = 0
    return hundred_digit-5
```
